# <font color=green> Manipulando os Dados





```
1) Do número cumulativo de mortes confirmadas ao longo do tempo;
2) Do número cumulativo de casos confirmados por mil habitantes ao longo do tempo;
3) Do número cumulativo de mortes confirmadas por mil habitantes ao longo do tempo.
```





````python
dados.columns = ["data",	"dia",	"mes",	"ano",	"casos",	"mortes",	"pais",	"Id",	"Code",	"popData2018"]
````

````python
dados
````

````python
dados2 = dados.set_index("data")
dados2 = dados2.sort_index()
dados2.head(5)
````

````python
numero_total_de_mortes = dados2['mortes'].sum()
numero_total_de_casos = dados2['casos'].sum()
print("O número total de casos é " , numero_total_de_casos)
print("O número total de mortes é " , numero_total_de_mortes)
````

````python
dados3 = dados2
````

````python
dados3.head(5)
````

````python
dados3["CUM_SUM_MORTES"] = dados2.groupby("pais")['mortes'].cumsum()
dados3["CUM_SUM_CASOS"] = dados2.groupby("pais")['casos'].cumsum()
dados3["CUM_SUM_CASOS_POR_MIL_HAB"] = dados2.groupby("pais")['casos'].cumsum()/1000
dados3["CUM_SUM_MORTES_POR_MIL_HAB"] = dados2.groupby("pais")['mortes'].cumsum()/1000
````

````python
dados3.head(40)
````

# <font color=blue> **Gráficos Exercícios**

````python
dados4 = dados3.pivot(columns="pais")
dados4['CUM_SUM_MORTES'].plot(title='Acumulo de Número de Mortes ao longo do Tempo', figsize=(20,10))
````

````python
dados4['CUM_SUM_CASOS'].plot(title='Acumulo de Número de Casos ao longo do Tempo', figsize=(20,10))
````

````python
dados4['CUM_SUM_CASOS_POR_MIL_HAB'].plot(title='Acumulo de Número de Casos por 1000 Habitantes ao longo do Tempo ', figsize=(20,10))
````

````python
dados4['CUM_SUM_MORTES_POR_MIL_HAB'].plot(title='Acumulo de Número de Mortes por 1000 Habitantes ao longo do Tempo ', figsize=(20,10))
````
